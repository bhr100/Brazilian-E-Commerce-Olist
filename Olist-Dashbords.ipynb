```python
# Import Libraries
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import dash_bootstrap_components as dbc
from dash import html ,dcc ,Dash
from dash.dependencies import Input ,Output 
from jupyter_dash import JupyterDash
from dash.exceptions import PreventUpdate
import dash_mantine_components as dmc  
from dash_bootstrap_templates import load_figure_template
```

```python
olist = pd.read_csv('E:\\portfolio\\Brazilian E-Commerce Public Dataset by Olist\\Data\\Olist.csv')
pd.set_option('display.max_column',None)
olist
```

```python
def Growth(data , column_name):   
                Growth = []
                for index , col in data.iterrows():         
                        try:       
                                Growth.append(round((data[column_name][index  + 1 ] - col[column_name]) / data[column_name][index  + 1],2)) # (Current - Pervious) / Current
                        except KeyError:                  
                                Growth.insert(0,0)  # To adjust the length with the original data         
                return Growth 
        
def Churn(data , column_name):   
                Churn = []
                for index , col in data.iterrows():         
                        try:       
                                Churn.append(round((col[column_name] - data[column_name][index  + 1 ]) / col[column_name] ,2))  # (previous - current) / previous
                        except KeyError:                  
                                Churn.insert(0,0)  # To adjust the length with the original data         
                return Churn 

df_line = olist.groupby(olist.order_purchase_timestamp.astype(np.datetime64).dt.date).agg(Total_Orders = ('order_id','nunique'),
                                                                                Total_Customers = ('customer_unique_id','nunique'),
                                                                                Total_Revenue = ('total_revenue','sum'),
                                                                                Total_Transaction =('order_id','count'))\
                                                                                .reset_index()\
                                                                                .assign(Average_Order_Value = lambda x : round(x['Total_Orders'] / x['Total_Transaction'],3),
                                                                                        Average_Revenue_Per_Order = lambda x : round(x['Total_Revenue']/x['Total_Orders']),
                                                                                        Cumulative_Total_Orders = lambda x : x['Total_Orders'].rolling(window=len(x) , min_periods=1).sum(),
                                                                                        Cumulative_Total_Revenue = lambda x : x['Total_Revenue'].rolling(window=len(x) , min_periods=1).sum(),
                                                                                        Percentage_Of_Cumulative_Orders = lambda x : (x['Total_Orders'] / x['Total_Orders'].sum()).rolling(window=len(x) , min_periods=1).sum(),
                                                                                        Percentage_Of_Cumulative_Revenue = lambda x : (x['Total_Revenue'] / x['Total_Revenue'].sum()).rolling(window=len(x) , min_periods=1).sum(),
                                                                                        Growth_Of_Total_Orders = lambda x : Growth(x , 'Total_Orders'),
                                                                                        Growth_Of_Total_Revenue = lambda x : Growth(x , 'Total_Revenue'),
                                                                                        Churn_Of_Total_Revenue = lambda x : Churn(x , 'Total_Revenue'),
                                                                                        Churn_Of_Total_Customers = lambda x : Churn(x , 'Total_Customers'),
                                                                                        Rolling_Average_Of_Total_Orders = lambda x : x['Total_Orders'].rolling(window=12).mean().round() ,
                                                                                        Rolling_Average_Of_Total_Revenue = lambda x : x['Total_Revenue'].rolling(window=12).mean().round())\
                                                                                .rename(columns={'order_purchase_timestamp':'Date'})\
                                                                                .sort_values('Date')
df_line.columns = df_line.columns.str.replace('_',' ')
df_line
```

```python
# Create Dash App
with open('H:/CS50/Portfolio Projects/Northwind Traders/assets/reset.css','w') as file:
        file.write('body{background-color:rgb(34, 34, 34);}\n')
        file.close()

dbc_css = "/assets/reset.css"

app = Dash(__name__,external_stylesheets=[dbc.themes.DARKLY,dbc_css] )


load_figure_template('DARKLY')

app.layout = dbc.Container(
                 fluid=True,
                 style={'backgroundColor':'rgb(34, 34, 34)','color':'white','border':'rgb(34, 34, 34)'},
                        children=[
                                dmc.Grid(
                                       children=[
                                                html.H1("🌎Brazilian🇧🇷 🛒E-Commerce ⭕list Dashboard",style={'text-align':'center'}),
                                                 dmc.Col([
                                                        dmc.Stack(
                                                                [
                                                                dcc.Dropdown(
                                                                id='dropdown-x',
                                                                options=[{'label':html.Span(col.replace("_"," ") if "_" in col else col , style={'color':'black'}),'value':col } for col in olist.select_dtypes('number').columns[1:]],
                                                                value='price',
                                                                placeholder='Select X column... ',
                                                                className='dbc',
                                                                ),  
                                                                dcc.Dropdown(
                                                                id='dropdown-y',
                                                                options=[{'label':html.Span(col.replace("_"," ") if "_" in col else col , style={'color':'black'}),'value':col } for col in olist.select_dtypes('number').columns[1:]],
                                                                placeholder='Select Y column... ',
                                                                value='freight_value',
                                                                className='dbc',
                                                                )
                                                                ],
                                                                align="stretch",
                                                                justify="flex-start",
                                                                spacing='xs'),
                                                        ],span=12),
                                                dmc.Col([
                                                        dcc.Graph(id='scatter-cross-filter',
                                                            className='dbc',
                                                            hoverData={'points':[{'customdata':['sao paulo']}]})
                                                       ],span=5),
                                               dmc.Col([dcc.Graph(id='hist-year-revenue',className='dbc')],span=7),
                                               html.Br(), 
                                               dmc.Col([
                                                       dcc.Dropdown(id='dropdown-original',
                                                         options=[{'label':html.Span(col.replace("_"," ") if "_" in col else col , style={'color':'black'}),'value':col } for col in df_line.columns[1:5]],
                                                         value='Total Orders',
                                                         placeholder = 'Select Original Column...',
                                                         className='dbc')
                                                         ],span=6), 
                                               dmc.Col([
                                                       dcc.Dropdown(id='dropdown-Secondary',
                                                         options=[{'label':html.Span(col.replace("_"," ") if "_" in col else col , style={'color':'black'}),'value':col } for col in df_line.columns[5:]],
                                                         value='Cumulative Total Revenue',
                                                         placeholder = 'Select Secondary Column...',
                                                         className='dbc')
                                                         ],span=6),
                                               dmc.Col([dcc.Graph(id='line-viz',className='dbc')],span=12)                                       
                                               ]),
                                          
                         ])
          

###############################################################

@app.callback(Output('scatter-cross-filter','figure'),
              Input('dropdown-x','value'),
              Input('dropdown-y','value'))

def scatter_cross(x,y):
    
    scatter = px.scatter(olist,
            x=x,
            y=y,
            hover_name='customer_city',
            hover_data=['customer_city'],
            marginal_x="histogram", 
            marginal_y="rug",
            trendline='ols',
            trendline_color_override="lightgray",
            )\
    .update_layout(showlegend=False)
    
    return scatter

###############################################################   

@app.callback(Output('hist-year-revenue','figure'),
              Input('scatter-cross-filter','hoverData'))
def Update_Bar(hoverData):

    city_name = hoverData['points'][0]['customdata'][0]

    df = pd.concat([olist.query('customer_city == @city_name').groupby('Year')['total_revenue'].describe().reset_index(),
           olist.query('customer_city == @city_name').groupby('Year')['total_revenue'].sum().reset_index().drop('Year',axis=1)],axis=1)\
           .rename(columns={'count':'Total Transaction','mean':'Total Average','std':'Standard deviation','min':'Min','max':'Max','total_revenue':'Total Revenue'}).round(2)
    
    bar = px.bar(df,
        x='Year',
        y=df.columns[1:],
        barmode='group',
        # color_discrete_sequence=px.colors.sequential.gray,
        # text_auto=True,
        labels={'variable':'','value':''},
        title=f'Descripe Of Total Revenue by {city_name.title()} For Each Year. ' 
        )\
        .update_traces(textposition='outside',
                       marker={'line_width':1})\
        .update_layout(
                        font_size = 10,
                        font_family='verdana',
                        legend_orientation='h',
                        xaxis={'type':'category','showgrid':False},
                        # bargap=0.4
                        )
    return bar

###############################################################

@app.callback(Output('line-viz','figure'),
              Input('dropdown-original','value'),
              Input('dropdown-Secondary','value'),
 )

def Line(original , secondary):

        colors = ['wheat','seagreen','tomato','gold']

        line = make_subplots(specs=[[{"secondary_y": True}]])


        line.add_trace(go.Scatter( x =df_line.Date ,
                        y = df_line[original],
                        name=f'{original}',
                        hoverinfo='x+y',
                        marker={'color':'silver'},
                        line_shape='spline',
                        ),secondary_y=False)

        line.add_trace(go.Scatter( x =df_line.Date ,
                                y = df_line[secondary],
                                name=f'{secondary}',
                                hoverinfo='x+y',
                                marker={'color':'skyblue'},
                                line_shape='linear'
                                ),secondary_y=True)

        df_line['Quantiles'] = pd.cut(df_line[original],  bins= list(df_line[original].quantile([0,.25,.5,.75,1])) , labels=['25%','50%','75%','100%']).fillna(value='25%')

        for num , quantiles in enumerate(df_line['Quantiles'].unique()):

                Quantiles_original = list(df_line[original].quantile([.25,.5,.75,1]))

                Quantiles_secondary = list(df_line[secondary].quantile([.25,.5,.75,1]))

                total_original = df_line[df_line['Quantiles'] == quantiles][original].sum()


                line.add_hrect(
                        y0=df_line[df_line['Quantiles'] == quantiles].sort_values(by=original)[original].min(),
                        y1=df_line[df_line['Quantiles'] == quantiles].sort_values(by=original)[original].max(),
                        line_width=.25,
                        fillcolor=colors[num],
                        opacity=0.25
                        ) 
                
                text = f'<b>{original} = {int(total_original)} | {secondary} = {int(Quantiles_secondary[num])} | {quantiles} <= {round(Quantiles_original[num])}<b>'
                y = [0.08,0.125,0.175,0.95]
                line.add_annotation(
                        x=0.001,
                        y=y[num],
                        text=text,
                        xref="paper",
                        yref="paper",
                        showarrow=False,
                        font_size=10,
                        font_color='snow'
                        
                )
        
        line.update_xaxes(rangeslider_visible=True)\
                .update_yaxes(visible=False, fixedrange=True)\
                .update_traces(line_width = 2)\
                .update_layout(legend_orientation='h',legend_x=0.7,height=750)
        


        return line                                                                               
                                                                            

if __name__ == '__main__':
    
    
    app.run_server(debug=False)  
```
